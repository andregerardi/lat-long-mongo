#### verifica os dados inseridos

```python
# lista bancos (exclui system dbs)
user_dbs = [d for d in client.list_database_names() if d not in ('admin', 'local', 'config')]
print("Databases:", user_dbs)

# escolhe o primeiro banco de usuário encontrado (ou altere para o nome desejado)
if not user_dbs:
    print("Nenhum banco de usuário encontrado.")
else:
    db = client[user_dbs[0]]
    print("Usando DB:", db.name)
    cols = db.list_collection_names()
    print("Collections:", cols)

    # para cada coleção, mostra contagem, um documento de exemplo e os primeiros 10 documentos
    for coll_name in cols:
        coll = db[coll_name]
        print(f"\nCollection: {coll_name}")
        print("Count:", coll.count_documents({}))
        print("Sample doc:", coll.find_one())
        print("First 10 docs:")
        for doc in coll.find().limit(10):
            print(doc)
```
